# Groupby on multiple columns

```python
#how many calls were made in each month under each/every category

group_three = data.groupby(['month', 'item'])['date'].count()

print(group_three)
```

```python
data.head()
```

#aggregation - agg()

```python
group_four = data.groupby(['month', 'item']).agg(
    
        {
            'date' : 'count',
            'duration' : [sum, min, max],
            'network' : 'first'
        } 
) 
print(group_four)
```

```python
import numpy as np
```

```python
x = np.arange(1, 21)
y = np.random.randint(20, size=20)

print(x)
print(y)
```

```python
plt.plot(x, y, 'gs--')

plt.xlabel("x coordinates", size=12)
plt.ylabel("y coordinates", size=12)
plt.title("Sample Plot")

for inx in range(len(x)):
  plt.text(x[inx], y[inx], str(x[inx]) +','+str(y[inx])  )
#plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,8))

plt.plot(x, y, 'gs--')

plt.xlabel("x coordinates", size=12)
plt.ylabel("y coordinates", size=12)
plt.title("Sample Plot")

for inx in range(len(x)):
  plt.text(x[inx], y[inx], str(x[inx]) +','+str(y[inx])  )

plt.xticks(x)
plt.yticks(x)

plt.xlim(0, 25)
plt.ylim(0, 20)
plt.grid()
```

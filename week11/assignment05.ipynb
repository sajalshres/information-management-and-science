```python
# imports
import pandas as pd
```

```python
# load data
data = pd.read_csv("Employees.csv")
data.head()
```

##### 1. Write query to find how many null-values are there in each column.

```python
for column in data.columns:
     print(column, "column has", data[column].isnull().sum(), "null values")
```

##### 2. Write query to find distinct values and their counts in each column.

```python
for column in data.columns:
    print(column, len(data[column].unique()))
    print(data[column].unique())
```

##### 3. Drop the rows where the employees are null/empty.

```python
data.dropna()
```

##### 4. Fill in the null/empty gender values with Unknown value.

```python
data = data.fillna({"Gender": "Unknown"})

print(data.Gender.unique())
```

##### 5. Fill in the null/empty Team values with some new department value.

```python
data = data.fillna({"Team":"Research and Development"})

print(data.Team.unique())
```

##### 6. For the columns having empty-spaces in their names, replace them with underscore _. For example, change the First Name column to First_Name and so on.

```python
new_columns = { column:"_".join(column.split(" ")) for column in data.columns }

data = data.rename(columns=new_columns)

print(data.columns)
```

##### 7. Change the format of each date from (mm/dd/yyyy) to (dd/mm/yyyy).

```python
def change_format(data):
    month, day, year = data.split("/")
    return f"{day}/{month}/{year}"

# data["Start_Date"] = data["Start_Date"].apply(change_format)

# or use built-ins
data["Start_Date"] = pd.to_datetime(data["Start_Date"])
data["Start_Date"] = data["Start_Date"].dt.strftime('%d/%m/%Y')

print(data.head())
```

##### 8. Round the bonus values to the next ceiling value, and change the data-type from float to int.

```python
import math

def to_ceil(value):
    return int(math.ceil(value))

data["Bonus_%"] = data["Bonus_%"].map(to_ceil)

print(data.head())

print(data.dtypes)
```

##### 9. For the Last_Login Column , remove the AM/PM strings, and convert it to 24-hour format.

```python
def to_24_hour(time):
    new_time, pm_am = time.split(" ")
    hour, minutes = new_time.split(":")

    if pm_am.lower() == "pm" and hour != "12":
        hour = str(int(hour) + 12)
    elif hour == "12":
            hour = "00"
    
    if (len(hour)) == 1:
        hour = "0" + hour

    return f"{hour}:{minutes}"


test_times = ["12:42 PM", "2:20 PM", "10:00 PM", "12:30 AM", "02:45 AM", "09:45 AM"]

for test_time in test_times:
    print(to_24_hour(test_time))


data["Last_Login_Time"] = data["Last_Login_Time"].map(to_24_hour)

# or use built-in
# data["Last_Login_Time"] = pd.to_datetime(data["Last_Login_Time"]).dt.strftime("%H:%M")

data.head()
```

##### 10. For the Senior_Management column, use forward fill technique to fill in the missing values.

```python
data["Senior_Management"] = data["Senior_Management"].fillna(method="ffill")

print(data.Senior_Management.isnull().sum())
```

##### 11. For the First_Name column, remove or drop rows having duplicate values. That is, no two employees can have same name (lets assume that it might later act as a primary-key).

```python
data["First_Name"].drop_duplicates()
```

##### 12. Convert the salary values from numbers to strings. Add a currency symbol (dollar, pound etc.)
at the beginning of the string.

```python
data["Salary"] = data["Salary"].astype(str)

def add_currency_symbol(salary):
    return "$" + salary

data["Salary"] = data["Salary"].map(add_currency_symbol)

data.head()
```

```python
#map, lambda

nums = [1,2,3,4]

squared = []

for num in nums :
  squared.append(num ** 2)

print(squared)
```

```python
def square_num(num):
  return (num ** 2)
```

```python
m = map(square_num, nums)
#print(m)
output = list(m)
print(output)
```

```python
m = map(lambda x : (x**2), nums)
output = list(m)
print(output)
```

```python
#list-comprehension
```

```python
nums = [1,2,3,4]

squared = []

for num in nums :
  if(num % 2 == 0):
    squared.append(num ** 2)
  else:
    squared.append(num)

print(squared)
```

```python
#list-comprehension

squared = []
nums = [1,2,3,4]

squared = [  (val**2) for val in nums  ]
print(squared)
```

```python
squared = []
nums = [1,2,3,4]

squared = [  (val**2) for val in nums  if (val % 2 == 0)]
print(squared)
```

```python
squared = []
nums = [1,2,3,4]

squared = [  (val**2) if (val % 2 == 0) else (val**3) for val in nums  ]
print(squared)
```

```python
#map
nums = [1,2,3,4]
m = map(lambda x : (x**2), nums)
output = list(m)
print(output)
```

```python
#filter
nums = [1,2,3,4]

f = filter(lambda x : x%2 != 0,  nums)
#print(f)
output = list(f)
print(output)
```

```python
#list-comprehension

my_customers = ['Sam', 'Mark', None, 'Sara', None, 'Tim', 'Kim']

new_list = [ c for c in my_customers if(c is not None)]
print(new_list)
```

```python
my_customers = ['Sam', 'Mark', None, 'Sara', None, 'Tim', 'Kim']

new_list = [ c if(c is not None) else '' for c in my_customers ]
print(new_list)
```

```python
#map w.r.t. dictionary

data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon','Pastrami', 'corned beef', 'Bacon','pastrami', 'honey ham', 'nova lox'],'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

print(data)
```

```python
#outliers
import numpy as np
```

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.head()
```

```python
col2 = data[2]
col2
```

```python
col2[col2 > 2.5]
```

```python
# valid values : (-2.5 to 2.5)  ; 
# all other values : outliers
```

```python
col2[ np.abs(col2) > 2.5]
```

```python
data[ np.abs(data) > 2.5]        #all - all the values in the row should be above the threshold
```

```python
data[ ( np.abs(data) > 2.5).any(1) ]     #any - atleast one value 
```

```python
#fix the outliers (cap the outliers) to some threshold value 

data[ np.abs(data) > 2.5] = np.sign(data) * 2.5
```
